## Data Analysis

### Top 5 most popular artists

```python
top_five_artists = df.groupby("artist").count().sort_values(by="song_title", ascending=False)["song_title"][:5]
top_five_artists
```

```python
top_five_artists.plot.barh()
plt.show()
```

### Top 5 loudest tracks


```python
top_five_loudest = df[["loudness","song_title"]].sort_values(by="loudness", ascending=True)[:5]
top_five_loudest
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x="loudness", y="song_title", data=top_five_loudest)
plt.title("Top 5 loudes tracks")
plt.show()
```

### Artist with the most danceability song

```python
top_five_artists_danceable_songs = df[["danceability","song_title","artist"]].sort_values(by="danceability", ascending=False)[:5]
top_five_artists_danceable_songs
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x="danceability", y="artist", data=top_five_artists_danceable_songs)
plt.title("Artists With Most Danceability")
```

### Top 10 instrumentalness tracks

```python
top_ten_inst_tracks = df[["instrumentalness", "song_title", "artist"]].sort_values(by="instrumentalness", ascending=False)[:5]
top_ten_inst_tracks
```

```python
plt.figure(figsize=(12,7))
plt.pie(x="instrumentalness", data =top_ten_inst_tracks, autopct='%1.2f%%', labels=top_ten_inst_tracks.song_title)
plt.show()
```

### Multiple feature plots

```python
interest_feat_col= ["tempo", "loudness", "acousticness", "danceability",
                    "duration_ms", "energy", "instrumentalness", "liveness", "speechiness", "valence"]
```

```python
for feature_col in interest_feat_col:
    pos_data = df[df["target"]== 1][feature_col]
    neg_data = df[df["target"]== 0][feature_col]
    
    plt.figure(figsize=(12,7))
    
    sns.displot(pos_data, bins=30, label="Positive", color="green")
    sns.displot(neg_data, bins=30, label="Negative", color="red")
    
    plt.legend(loc="upper right")
    plt.title(f"Positive and Negatie Histogram Plot For {feature_col}")
    plt.show()
```
